Get place in the tournament table.

```python
def result(HG, AG):
    if HG > AG:
        return (3, 0)
    if HG < AG:
        return (0, 3)
    return (1, 1)
```

```python
tables = dict()
teams_L_S = dict()

for L in matches['league'].unique():
    for S in matches['season'].unique():
        tmp_df = matches[
            (matches['league'] == L)
            &
            (matches['season'] == S)
        ]

        teams = list(tmp_df['home_team'].unique())
        teams.sort(reverse=True)

        standings = []
        for team in teams:
            standings.append([0, 0, teams.index(team)])

        tables[(L, S)] = standings
        teams_L_S[(L, S)] = teams
```

```python
teams_L_S[('EPL', '2022/2023')]
```

```python
home_standings = []
away_standings = []
home_points = []
away_points = []

for match_i in range(len(matches)):
    HG, AG = int(matches.iloc[match_i]['home_score']), int(matches.iloc[match_i]['away_score'])
    points = result(HG, AG)

    home, away = matches.iloc[match_i]['home_team'], matches.iloc[match_i]['away_team']
    L, S = matches.iloc[match_i]['league'], matches.iloc[match_i]['season']

    teams = teams_L_S[(L, S)]
    standings = tables[(L, S)]
    standings.sort(reverse=True)

    for i in range(len(standings)):
        if home == teams[standings[i][2]]:
            home_standings.append(i + 1)
            home_points.append(standings[i][0])
            standings[i] = [
                standings[i][0] + points[0],
                standings[i][1] + HG - AG,
                standings[i][2]
            ]
            break
    for i in range(len(standings)):
        if away == teams[standings[i][2]]:
            away_standings.append(i + 1)
            away_points.append(standings[i][0])
            standings[i] = [
                standings[i][0] + points[1],
                standings[i][1] + AG - HG,
                standings[i][2]
            ]
            break

    tables[(L, S)] = standings

matches['home_standings'] = home_standings
matches['away_standings'] = away_standings
matches['home_points'] = home_points
matches['away_points'] = away_points
```

Get stats for the last 4 matches (both teams), last 2 home matches (home team), last 2 away matches (away team):

1. Points
2. Goals
3. Goals against
4. Mean xG
5. Mean xG against

```python
team_stats = dict()

for team in matches['home_team'].unique():
    team_stats[team] = {
        'points': [np.nan, np.nan, np.nan, np.nan],
        'home_points': [np.nan, np.nan],
        'away_points': [np.nan, np.nan],
        'goals': [np.nan, np.nan, np.nan, np.nan],
        'home_goals': [np.nan, np.nan],
        'away_goals': [np.nan, np.nan],
        'goals_against': [np.nan, np.nan, np.nan, np.nan],
        'home_goals_against': [np.nan, np.nan],
        'away_goals_against': [np.nan, np.nan],
        'xg': [np.nan, np.nan, np.nan, np.nan],
        'home_xg': [np.nan, np.nan],
        'away_xg': [np.nan, np.nan],
        'xg_against': [np.nan, np.nan, np.nan, np.nan],
        'home_xg_against': [np.nan, np.nan],
        'away_xg_against': [np.nan, np.nan]
    }
```

```python
home_points_last_4_matches = []
away_points_last_4_matches = []
home_points_last_2_home_matches = []
away_points_last_2_away_matches = []

home_goals_last_4_matches = []
away_goals_last_4_matches = []
home_goals_last_2_home_matches = []
away_goals_last_2_away_matches = []

home_goals_against_last_4_matches = []
away_goals_against_last_4_matches = []
home_goals_against_last_2_home_matches = []
away_goals_against_last_2_away_matches = []

home_mean_xg_last_4_matches = []
away_mean_xg_last_4_matches = []
home_mean_xg_last_2_home_matches = []
away_mean_xg_last_2_away_matches = []

home_mean_xg_against_last_4_matches = []
away_mean_xg_against_last_4_matches = []
home_mean_xg_against_last_2_home_matches = []
away_mean_xg_against_last_2_away_matches = []

for match_i in range(len(matches)):
    home, away = matches.iloc[match_i]['home_team'], matches.iloc[match_i]['away_team']
    date = matches.iloc[match_i]['datetime']

    home_goals, away_goals = int(matches.iloc[match_i]['home_score']), int(matches.iloc[match_i]['away_score'])
    home_goals_against, away_goals_against = away_goals, home_goals
    home_points, away_points = result(home_goals, away_goals)

    home_xg, away_xg = float(matches.iloc[match_i]['home_xg']), float(matches.iloc[match_i]['away_xg'])
    home_xg_against, away_xg_against = away_xg, home_xg

    home_points_last_4_matches.append(np.sum(team_stats[home]['points']))
    away_points_last_4_matches.append(np.sum(team_stats[away]['points']))
    home_points_last_2_home_matches.append(np.sum(team_stats[home]['home_points']))
    away_points_last_2_away_matches.append(np.sum(team_stats[away]['away_points']))

    home_goals_last_4_matches.append(np.sum(team_stats[home]['goals']))
    away_goals_last_4_matches.append(np.sum(team_stats[away]['goals']))
    home_goals_last_2_home_matches.append(np.sum(team_stats[home]['home_goals']))
    away_goals_last_2_away_matches.append(np.sum(team_stats[away]['away_goals']))

    home_goals_against_last_4_matches.append(np.sum(team_stats[home]['goals_against']))
    away_goals_against_last_4_matches.append(np.sum(team_stats[away]['goals_against']))
    home_goals_against_last_2_home_matches.append(np.sum(team_stats[home]['home_goals_against']))
    away_goals_against_last_2_away_matches.append(np.sum(team_stats[away]['away_goals_against']))

    home_mean_xg_last_4_matches.append(np.mean(team_stats[home]['xg']))
    away_mean_xg_last_4_matches.append(np.mean(team_stats[away]['xg']))
    home_mean_xg_last_2_home_matches.append(np.mean(team_stats[home]['home_xg']))
    away_mean_xg_last_2_away_matches.append(np.mean(team_stats[away]['away_xg']))

    home_mean_xg_against_last_4_matches.append(np.mean(team_stats[home]['xg_against']))
    away_mean_xg_against_last_4_matches.append(np.mean(team_stats[away]['xg_against']))
    home_mean_xg_against_last_2_home_matches.append(np.mean(team_stats[home]['home_xg_against']))
    away_mean_xg_against_last_2_away_matches.append(np.mean(team_stats[away]['away_xg_against']))

    team_stats[home] = {
        'points': [home_points, team_stats[home]['points'][0], team_stats[home]['points'][1], team_stats[home]['points'][2]],
        'home_points': [home_points, team_stats[home]['home_points'][0]],
        'away_points': team_stats[home]['away_points'],
        'goals': [home_goals, team_stats[home]['goals'][0], team_stats[home]['goals'][1], team_stats[home]['goals'][2]],
        'home_goals': [home_goals, team_stats[home]['home_goals'][0]],
        'away_goals': team_stats[home]['away_goals'],
        'goals_against': [home_goals_against, team_stats[home]['goals_against'][0], team_stats[home]['goals_against'][1], team_stats[home]['goals_against'][2]],
        'home_goals_against': [home_goals_against, team_stats[home]['home_goals_against'][0]],
        'away_goals_against': team_stats[home]['away_goals_against'],
        'xg': [home_xg, team_stats[home]['xg'][0], team_stats[home]['xg'][1], team_stats[home]['xg'][2]],
        'home_xg': [home_xg, team_stats[home]['home_xg'][0]],
        'away_xg': team_stats[home]['away_xg'],
        'xg_against': [home_xg_against, team_stats[home]['xg_against'][0], team_stats[home]['xg_against'][1], team_stats[home]['xg_against'][2]],
        'home_xg_against': [home_xg_against, team_stats[home]['home_xg_against'][0]],
        'away_xg_against': team_stats[home]['away_xg_against']
    }

    team_stats[away] = {
        'points': [away_points, team_stats[away]['points'][0], team_stats[away]['points'][1], team_stats[away]['points'][2]],
        'away_points': [away_points, team_stats[away]['away_points'][0]],
        'home_points': team_stats[away]['home_points'],
        'goals': [away_goals, team_stats[away]['goals'][0], team_stats[away]['goals'][1], team_stats[away]['goals'][2]],
        'away_goals': [away_goals, team_stats[away]['away_goals'][0]],
        'home_goals': team_stats[away]['home_goals'],
        'goals_against': [away_goals_against, team_stats[away]['goals_against'][0], team_stats[away]['goals_against'][1], team_stats[away]['goals_against'][2]],
        'away_goals_against': [away_goals_against, team_stats[away]['away_goals_against'][0]],
        'home_goals_against': team_stats[away]['home_goals_against'],
        'xg': [away_xg, team_stats[away]['xg'][0], team_stats[away]['xg'][1], team_stats[away]['xg'][2]],
        'away_xg': [away_xg, team_stats[away]['away_xg'][0]],
        'home_xg': team_stats[away]['home_xg'],
        'xg_against': [away_xg_against, team_stats[away]['xg_against'][0], team_stats[away]['xg_against'][1], team_stats[away]['xg_against'][2]],
        'away_xg_against': [away_xg_against, team_stats[away]['away_xg_against'][0]],
        'home_xg_against': team_stats[away]['home_xg_against']
    }

matches['home_points_last_4_matches'] = home_points_last_4_matches
matches['away_points_last_4_matches'] = away_points_last_4_matches
matches['home_points_last_2_home_matches'] = home_points_last_2_home_matches
matches['away_points_last_2_away_matches'] = away_points_last_2_away_matches
matches['home_goals_last_4_matches'] = home_goals_last_4_matches
matches['away_goals_last_4_matches'] = away_goals_last_4_matches
matches['home_goals_last_2_home_matches'] = home_goals_last_2_home_matches
matches['away_goals_last_2_away_matches'] = away_goals_last_2_away_matches
matches['home_goals_against_last_4_matches'] = home_goals_against_last_4_matches
matches['away_goals_against_last_4_matches'] = away_goals_against_last_4_matches
matches['home_goals_against_last_2_home_matches'] = home_goals_against_last_2_home_matches
matches['away_goals_against_last_2_away_matches'] = away_goals_against_last_2_away_matches
matches['home_mean_xg_last_4_matches'] = home_mean_xg_last_4_matches
matches['away_mean_xg_last_4_matches'] = away_mean_xg_last_4_matches
matches['home_mean_xg_last_2_home_matches'] = home_mean_xg_last_2_home_matches
matches['away_mean_xg_last_2_away_matches'] = away_mean_xg_last_2_away_matches
matches['home_mean_xg_against_last_4_matches'] = home_mean_xg_against_last_4_matches
matches['away_mean_xg_against_last_4_matches'] = away_mean_xg_against_last_4_matches
matches['home_mean_xg_against_last_2_home_matches'] = home_mean_xg_against_last_2_home_matches
matches['away_mean_xg_against_last_2_away_matches'] = away_mean_xg_against_last_2_away_matches
```

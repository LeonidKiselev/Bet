All together.

```python
def get_xg(s):
    return s[s.find('<div class="progress-value">') + 28:s.find('<span')] + s[s.find('<span class="progress-value-decimal">') + 37:s.find('</span>')]
```

```python
def get_player_name(s):
    return s[s.find('<span class="player-name">') + 26:s.find('</span>')]
```

```python
def get_player_position(s):
    return s[s.find('<span class="player-position">') + 30:s.find('</span>')]
```

```python
def get_data(match):
    url = f'https://understat.com/match/{match}'

    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')

    title = str(soup.find('title'))

    tmp = title[title.find('<title>') + 7:title.find(' -')]
    home_team = tmp[:tmp.rfind(' ')]
    home_score = tmp[tmp.rfind(' ') + 1:]

    tmp = title[title.find('- ') + 2:title.find(' (')]
    away_score = tmp[:tmp.find(' ')]
    away_team = tmp[tmp.find(' ') + 1:]

    tmp = title[title.find(' (') + 2:title.rfind(' | ')]
    date = tmp[:tmp.find(') ')]
    league = tmp[tmp.find(' | ') + 3:tmp.find(' | ', tmp.find(' | ') + 1)]
    season = tmp[tmp.find(' | ', tmp.find(' | ') + 1) + 3:tmp.rfind(' | ', tmp.rfind(' | ') - 1)]

    stats = soup.find_all('div', class_='progress-value')

    tmp = str(stats[7])
    home_xg = get_xg(tmp)

    tmp = str(stats[8])
    away_xg = get_xg(tmp)

    d = {
        'home_team': home_team,
        'away_team': away_team,
        'date': date,
        'league': league,
        'season': season,
        'home_score': int(home_score),
        'away_score': int(away_score),
        'home_xg': float(home_xg),
        'away_xg': float(away_xg),
    }

    player_names_raw = soup.find_all('span', class_='player-name')
    player_names = [get_player_name(str(s)) for s in player_names_raw]

    player_positions_raw = soup.find_all('span', class_='player-position')
    player_positions = [get_player_position(str(s)) for s in player_positions_raw]

    border = player_positions.index('GK', 1)

    tmp_names = player_names[:border]
    tmp_positions = player_positions[:border]
    for i in range(len(tmp_names)):
        d[f'home_player_name_{i + 1}'] = tmp_names[i]
        d[f'home_player_position_{i + 1}'] = tmp_positions[i]
    for i in range(len(tmp_names), 16):
        d[f'home_player_name_{i + 1}'] = np.nan
        d[f'home_player_position_{i + 1}'] = np.nan

    tmp_names = player_names[border:]
    tmp_positions = player_positions[border:]
    for i in range(len(tmp_names)):
        d[f'away_player_name_{i + 1}'] = tmp_names[i]
        d[f'away_player_position_{i + 1}'] = tmp_positions[i]
    for i in range(len(tmp_names), 16):
        d[f'away_player_name_{i + 1}'] = np.nan
        d[f'away_player_position_{i + 1}'] = np.nan

    return d
```

```python
get_data(4749)
```

Part 1 to 24.

```python
part = 25
```

```python
from datetime import datetime

start = datetime.now()

data = []

for i in range(1000 * (part - 1), 1000 * part):
    try:
        d = get_data(i)
        data.append(d)
    except:
        continue

df = pd.DataFrame.from_dict(data)
df.to_csv(f'understat_data_part_{part}.csv')

finish = datetime.now()
print((finish - start).total_seconds())
```

Concatenation.

```python
dfs = []
for i in range(1, 25):
    dfs.append(pd.read_csv(f'understat_data_part_{i}.csv'))
    print(i, end=' ')

df = pd.concat(dfs, ignore_index=True)
df.to_csv('understat_data_full.csv')
```

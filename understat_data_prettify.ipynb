Calculating the number of matches in each season.

```python
leagues = list(data['league'].unique())
seasons = list(data['season'].unique())
```

```python
for l in leagues:
    for s in seasons:
        total = len(data[np.logical_and(
            data['league'] == l,
            data['season'] == s
        )])
        print(f'{l} {s}: {total} matches')
```

Add datetime column.

```python
import datetime
```

```python
month_to_num = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}
```

```python
def str_to_date(s):
    s_list = s.split(' ')

    m = month_to_num[s_list[0]]
    d = int(s_list[1])
    y = int(s_list[2])

    return datetime.date(y, m, d)
```

```python
data['datetime'] = data['date'].apply(str_to_date)
```

```python
data.head()
```

Sort by league and then by datetime.

```python
data = data.sort_values(by=['league', 'datetime'])
```
